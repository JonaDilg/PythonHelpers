```python
import csv

import numpy as np
from numpy import array as arr

import pandas
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import pylandau
```

```python
# goes to the file line by line, writes hits (est. amplitude > threshold) to new file 
# (reduce file-size by 98%)

# datapathIn = "/home/jona/DESY/analysis_TB/output/csv/184_full.csv"
datapathIn = "/home/jona/DESY/analysis_TB/output/csv/190_full.csv"

datapathOut = "/home/jona/DESY/analysis_TB/output/csv/190_reduced.csv"


N_lines_max = 10000000000
N_lines = 0
N_hits = 0
printIt = False

```

```python
# only write out every n-th event

writeEvery = 50

fileIn = open(datapathIn, "r")
fileOut = open(datapathOut, "w")

reader = csv.reader(fileIn)
line = reader.__next__()
print(line)
lines = 0
counter = 0

fileOut.write(",".join(line) + "\n")
for i, line in enumerate(reader):
    if(i%400000==0):
        print(int(i/4), "events read")
        
    if type(lines) == int:
        lines = arr([line])
    elif lines[0][0] == line[0]:
        lines = np.append(lines, [line], axis=0)
    else:
        for l in lines:
            if l[7] == "0":
                l[13] = 0
                l[14] = 0
            if (counter%writeEvery == 0):
                fileOut.write(",".join(l) + "\n")
        lines = arr([line])
        counter += 1

    if(i>=N_lines_max):
        break
    
print("Read ", i, " lines containing ", i/4, " events & wrote ", counter/writeEvery, " events")

fileIn.close()
fileOut.close()
```

```python
# only write out hits

fileIn = open(datapathIn, "r")
fileOut = open(datapathOut, "w")

reader = csv.reader(fileIn)
line = reader.__next__()
print(line)
lines = 0
counter = 0

fileOut.write(",".join(line) + "\n")
for i, line in enumerate(reader):
    if(i%250000==0):
        print(int(i/4), "events read")
    if type(lines) == int:
        lines = arr([line])
    elif lines[0][0] == line[0]:
        lines = np.append(lines, [line], axis=0)
    else:
        if np.sum(lines[:,7].astype(float)) > 0:
            if printIt:
                print("evt #", lines[0,0])
            for l in lines:
                if l[7] == "0":
                    l[13] = 0
                    l[14] = 0
                if printIt:
                    print("line:", l)
                fileOut.write(",".join(l) + "\n")
            N_hits += 1
        lines = arr([line])

    if(i>=N_lines_max):
        break
print("Read ", i, " lines containing ", i/4, " events & found", N_hits, "hits => purity =", N_hits/(i/4)*100, "%")

fileIn.close()
fileOut.close()
```
